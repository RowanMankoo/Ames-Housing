# 3 Data cleaning

Before we start the data cleaning process we shall create a merged dataset of the test and training data, so that any transformations we apply are applied to both the test and training set, in the interest of saving time. 

*If we were to create a model and deploy it we would create custom pipelines for the data. However, as we only need to predict one test set in this competition it is easier to merge the datasets into one.*

```python
data = pd.concat([data_train,data_test])
```

## 3.1 Missing values


```python
total = data.isnull().sum().sort_values(ascending=False).head(20)
percentage =  100*(total/data.shape[0])

missing_data = pd.DataFrame({'Total':total, 'Percentage':percentage})

missing_data.head(20)
```

There seems to be a lot of NA values for certain variables, which leads us to believe that there has not been a data entry mistake for these NA values. After referring to the data key we find that the NA values for the first 16 variables (PoolQC to BsmtFinType1) are because the house does not have what the variable is assessing. For example, an NA for PoolQC would mean the house does not have a pool. So instead of imputing these values for the mean or most frequent one we shall replace them with a "none" value.

## 3.2 Correcting NA labels

Unfortunately, we cannot impute GarageYrBlt for "none" or any other alternative as it is a numerical variable so we are forced to drop it ([we drop it later on](#3.4-Dropping-variables)). This isn't too detrimental to our model as all the other garage variables correlate with GarageYrBlt as we have seen in our [correlation map](#2.2.2-Correlations-of-features).     

```python
# train imputer
imp = SimpleImputer(strategy='constant', fill_value='none')
labels = ['MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'GarageCond',
              'GarageType', 'GarageFinish', 'GarageQual', 'BsmtExposure', 'BsmtFinType2',
              'BsmtFinType1', 'BsmtCond', 'BsmtQual', 'PoolQC']
imp.fit(data[labels])
transformation = imp.transform(data[labels])

# loop over columns to change them
for i, col_name in enumerate(labels):
    data[col_name] = transformation[:,i]
```

## 3.3 Adding features


```python
data['HasPool'] = data['PoolArea'].apply(lambda x: 0 if x == 0 else 1)
data['TotalSF'] = data['TotalBsmtSF'] + data['1stFlrSF'] + data['2ndFlrSF']
data['Total_Bathrooms'] = data['FullBath'] + 0.5*data['HalfBath'] + data['BsmtFullBath']+ 0.5*data['BsmtHalfBath']
```

## 3.4 Dropping variables

Due to NA values, we shall drop **PoolQC** and **LotFrontage** (Note that on the [heatmap](#2.2.2-Correlations-of-features) we can see that LotFrontage and LotArea has a correlation of 0.5 so we do not lose too much information by dropping it). We shall also drop **ID** as this is of no importance in prediction and **PoolArea** due to its heavy skew.                     


```python
data = data.drop(['PoolQC','LotFrontage', 'Id', 'GarageYrBlt', 'PoolArea'], axis=1)
```

## 3.5 Ordinal encoding

From reading the data key we notice that some of the categorical variables that may contain information in their ordering set. We would like to preserve this information, so we shall manually encode them.

```python
Ordinal_features = ['LotShape', 'Utilities', 'LandSlope', 'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure',
                   'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageFinish',
                   'GarageQual', 'GarageCond', 'Fence']

# take note of numerical features before we encode the categorical ones
numerical = data.dtypes[data.dtypes != "object"].index

# we create dictionaries for the encodings
qual_dict = dict(none=-2,Po=-1,Fa=0,TA=1,Gd=2,Ex=3)
for feature in ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual',
                'FireplaceQu','GarageQual','GarageCond','PoolQC']:
    locals()['{0}_dict'.format(feature)] = qual_dict
LotShape_dict = dict(Reg=-1,IR1=0,IR2=1,IR3=2)
Utilities_dict = dict(AllPub=-2,NoSewr=-1,NoSeWa=0,ELO=1)
LandSlope_dict = dict(Gtl=-1,Mod=0,Sev=1)
BsmtExposure_dict = dict(none=-2,No=-1,Mn=0,Av=1,Gd=2)
BsmtFinType1_dict = dict(none=-3,Unf=-2,LwQ=-1,Rec=0,BLQ=1,ALQ=2,GLQ=3)
BsmtFinType2_dict = dict(none=-3,Unf=-2,LwQ=-1,Rec=0,BLQ=1,ALQ=2,GLQ=3)
Functional_dict = dict(Typ=-4,Min1=-3,Min2=-2,Mod=-1,Maj1=0,Maj2=1,Sev=2,Sal=3)
GarageFinish_dict = dict(none=-1,Unf=0,RFn=1,Fin=2)
Fence_dict = dict(none=-2,MnWw=-1,GdWo=0,MnPrv=1,GdPrv=2)

# apply encodings
for feature in Ordinal_features:
    data[feature] = data[feature].map(locals()['{0}_dict'.format(feature)])
```

## 3.6 Imputing the rest of the values

The rest of the NA's are probably due to missing data entries so we shall impute them. First, we shall split the dataset into a categorical and numerical dataset, to use two different imputing strategies.

```python
# split dataset into a categorical and numerical one
data_num = data[numerical] 
data_cat = data.drop(numerical, axis=1)

# take note of numerical and categorical column names before we 
# convert data into numpy array
data_num_cols = data_num.columns
data_cat_cols = data_cat.columns

# fit imputers
imp_num = SimpleImputer(strategy='mean')
imp_cat = SimpleImputer(strategy='most_frequent')

imp_num.fit(data_num)
imp_cat.fit(data_cat)

# transform data
data_num = imp_num.transform(data_num)
data_cat = imp_cat.transform(data_cat)
```

## 3.7 Spliting of data

Due to the nature of the models we require two different datasets, each requiring different pre-processing steps.



### 3.7.1 Decision tree train data

At this stage, the data is almost ready to be fed into any Decision tree-based models that we may use. The last thing we need to do is encode the categorical features. For decision tree models the best encoding strategy is to use an ordinal encoder instead of OneHotEncoding. As if the decision tree decides to split on one of the dummy variables the gain in purity per split is very marginal. However, for linear regression models, it would be best to use OneHotEncoding.

```python
# encode categorical features
enc1 = OrdinalEncoder()
enc1.fit(data_cat)
data_cat_rf = enc1.transform(data_cat)

# create decision tree dataset
X_RF = np.concatenate((data_num, data_cat_rf), axis=1)
X_train_RF = X_RF[:data_train.shape[0],]
X_test_RF = X_RF[data_train.shape[0]:,]
```

### 3.7.2 Linear regression train data


Recall from the [distribution of our features](#2.2.1-Distributions-of-features) some of them were skewed, we shall now check the extend of the skew and transform all features with a skew of more than 0.8 using a log transformation. This transformation helps our model become more robust to outliers in the data. 


```python
# function to retrieve skewed columns
def SkewedCols(dataset,skew_threshold):
    array = []
    for i in range(dataset.shape[1]):
         if abs(skew(dataset[:,i]))>skew_threshold:
            array.append(i)
    return array
    
skew_cols = SkewedCols(data_num,skew_threshold=0.8)

def SkewTable(dataset, skew_cols):
    
    # get skew values before transformation
    skew_before = np.empty((0,0))
    for i in range(dataset.shape[1]):
        skew_before = np.append(skew_before,skew(dataset[:,i]))
    
    # apply boxcox transformation
    for col_num in skew_cols:
        dataset[:,col_num] = boxcox1p(dataset[:,col_num], 0.15)
    
    # get skew values after transformation
    skew_after = np.empty((0,0))
    for i in range(dataset.shape[1]):
        skew_after = np.append(skew_after,skew(dataset[:,i]))
        
    
    return pd.DataFrame({'feature':data_num_cols,'Skew Before':skew_before,
                         'Skew After':skew_after})
    

SkewTable(data_num, skew_cols)
```

```python
# standardise the data
scl = StandardScaler()
data_num = scl.fit_transform(data_num)
```

Recall that some of our features have already been ordinally encoded ([Section 3.5](#3.5-Ordinal-encoding)), so we only need to one hot encode a subset of our categorical features.

```python
# find features which need OneHot encoding
OneHot_features = []
for i in data_cat_cols:
    if i not in Ordinal_features:
        OneHot_features.append(i)

# find positions of these features in numpy array
OneHot_positions =[]
for i in OneHot_features:
    for position,j in enumerate(data_cat_cols):
        if i==j:
            OneHot_positions.append(position)
            
# encode categorical features
enc2 = OneHotEncoder(sparse=False)
OneHot = enc2.fit_transform(data_cat[:,(OneHot_positions)])
temp_data_cat = np.delete(data_cat,OneHot_positions,axis=1)
```

```python
# create linear regression dataset
data_cat_LR = np.append(temp_data_cat,OneHot,axis=1)
X_LR = np.concatenate((data_num, data_cat_LR), axis=1)

X_train_LR = X_LR[:data_train.shape[0],]
X_test_LR = X_LR[data_train.shape[0]:,]
```

# 4 Model

Our final model will be an averaged prediction of 6 different models. To get a more accurate average prediction we shall average predictions with a lower correlation. To achieve this, we use models which make different assumptions about the data. The models in use are:
 
- XGBoostRegressor
- Elasticnet Regression
- GradientBoostingRegressor
- Lasso regression
- SVR (support vector regression)
- Ridge regression

## 4.1 Model training

We first fit our models using hyperparameters found via grid search cross-validation.

```python
regr = xgb.XGBRegressor(colsample_bytree=0.2, gamma=0.0, learning_rate=0.01, max_depth=4,
                        min_child_weight=1.5, n_estimators=7200, reg_alpha=0.9, reg_lambda=0.6, 
                        subsample=0.2).fit(X_train_RF,y)
en = ElasticNet(alpha =0.0005, l1_ratio=0.9, max_iter=10000).fit(X_train_LR,y_transformed)
gbr = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05, max_depth=4, max_features='sqrt', 
                                min_samples_leaf=15, min_samples_split=10, loss='huber').fit(X_train_RF,y)
la = Lasso(alpha=0.0005, max_iter=10000).fit(X_train_LR, y_transformed)
sv = SVR(C= 20, epsilon= 0.008, gamma=0.0003).fit(X_train_LR,y_transformed)
rid = Ridge(alpha=20).fit(X_train_LR,y_transformed)
```

## 4.2 Model evaluation

Now we evaluate our models using the RMSE metric via cross-validation. Note that we evaluate our decision tree-based models on y_transformed, even though when we fit these models we plan to fit them on y. If we evaluated on y then the RMSE would be a lot larger and we could not compare to that of the linear regression-based models.

```python
regr_score = np.sqrt(-cross_val_score(xgb.XGBRegressor(colsample_bytree=0.2, gamma=0.0, learning_rate=0.01, max_depth=4,
                        min_child_weight=1.5, n_estimators=7200, reg_alpha=0.9, reg_lambda=0.6, 
                        subsample=0.2), X_train_RF,y_transformed, scoring='neg_mean_squared_error')).mean()
en_score = np.sqrt(-cross_val_score(ElasticNet(alpha =0.0005, l1_ratio=0.9, max_iter=10000), X_train_LR,
                                    y_transformed, scoring='neg_mean_squared_error')).mean()
gbr_score = np.sqrt(-cross_val_score(GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05, max_depth=4,
                                                              max_features='sqrt', min_samples_leaf=15,
                                                              min_samples_split=10, loss='huber'), X_train_RF,
                                    y_transformed, scoring='neg_mean_squared_error')).mean()
la_score = np.sqrt(-cross_val_score(Lasso(alpha=0.0005, max_iter=10000), X_train_LR,
                                    y_transformed, scoring='neg_mean_squared_error')).mean()
sv_score = np.sqrt(-cross_val_score(SVR(C= 20, epsilon= 0.008, gamma=0.0003), X_train_LR,
                                    y_transformed, scoring='neg_mean_squared_error')).mean()
rid_score = np.sqrt(-cross_val_score(Ridge(alpha=20), X_train_LR, y_transformed,scoring='neg_mean_squared_error')).mean()
```

```python
names=['XGBRegressor', 'ElasticNet', 'GradientBoostingRegression', 'Lasso', 'SVR', 'Ridge']
Scores = [regr_score, en_score, gbr_score, la_score, sv_score, rid_score]
Models = pd.DataFrame(dict(Names=names, RMSE=Scores))
Models
```

## 4.3 Predictions


```python
pred1 = np.expm1(en.predict(X_test_LR))
pred2 = regr.predict(X_test_RF)
pred3 = gbr.predict(X_test_RF)
pred4 = np.expm1(la.predict(X_test_LR))
pred5 = np.expm1(sv.predict(X_test_LR))
pred6 = np.expm1(rid.predict(X_test_LR))

predictions = (1/6)*(pred1+pred2+pred3+pred4+pred5+pred6)
Id = data_test['Id'].ravel()
```

```python
# create submission file
my_submission = pd.DataFrame({'Id':Id, 'SalePrice':predictions})
my_submission.to_csv('Submission.csv', index=False)
```
